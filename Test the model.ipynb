```python
import pandas as pd
import pickle
from sklearn.preprocessing import StandardScaler


test_data_file_name='test__comments_data_set.csv'
train_data_file_name='train__comments_data_set.csv'
model_filename='comments_prediction_model.sav'

test=pd.read_csv(test_data_file_name)
train=pd.read_csv(train_data_file_name)

del test['Unnamed: 0']
del train['Unnamed: 0']

loaded_model = pickle.load(open(model_filename, 'rb'))
```

```python
x_test_data=test[test.columns[:-1]]
y_test_data=test[test.columns[-1]]
x_train_data=train[train.columns[:-1]]
y_train_data=train[train.columns[-1]]
```

```python
sc_X = StandardScaler()
x_train = sc_X.fit_transform(x_train_data)
x_test = sc_X.transform(x_test_data)

sc_Y = StandardScaler()
y_train=y_train_data.values.reshape(-1,1)
sc_Y.fit_transform(y_train)
y_train=sc_Y.transform(y_train)
y_test=y_test_data.values.reshape(-1,1)
y_test=sc_Y.transform(y_test)
```

```python
pred = loaded_model.predict(x_test)
predicated_value = sc_Y.inverse_transform(pred.reshape(-1,1)).reshape(-1)
```

```python
x_test_data['pred_value']=predicated_value
x_test_data
```
